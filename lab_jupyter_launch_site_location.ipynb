Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
distance_coastline = calculate_distance(28.563197, -80.576820, 28.56399, -80.56803)
print(distance_coastline)
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )
from folium.features import DivIcon

# Coordinates of the launch site
launch_site_lat = 28.563197
launch_site_lon = -80.576820

# Coordinates of the closest coastline
coastline_lat = 28.56399
coastline_lon = -80.56803

# Calculate the distance between the launch site and the coastline
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Create a Marker for the closest coastline point
distance_marker = folium.Marker(
    location=[coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(150, 36),  
        icon_anchor=(0, 0),   # Anchor point for icon
        html='<div style="font-size: 12pt; color: #f0edd8;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
    )
)

# Adding distance marker to the map
site_map.add_child(distance_marker)
site_map.save('map4_distance.html')

# Display the map
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
distance_highway = calculate_distance(28.563197, -80.576820, 28.56283, -80.57044)
print(distance_highway)
```

```python
# Coordinates
launch_site_lat, launch_site_lon = 28.563197, -80.576820
highway_lat, highway_lon = 28.56283, -80.57044
rail_lat, rail_lon = 28.57135, -80.58535
city_lat, city_lon = 28.34354, -80.6617

# Calculate distances using calculate_distance() function
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)
distance_rail = calculate_distance(launch_site_lat, launch_site_lon, rail_lat, rail_lon)
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)

# Print distances for verification
print(f"Distance to Highway: {distance_highway:.2f} km")
print(f"Distance to Railway: {distance_rail:.2f} km")
print(f"Distance to City: {distance_city:.2f} km")

# Create markers for highway, railway, and city
highway_marker = folium.Marker(
    location=[highway_lat, highway_lon],
    icon=DivIcon(
        icon_size=(150, 36),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12pt; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_highway),
    )
)

rail_marker = folium.Marker(
    location=[rail_lat, rail_lon],
    icon=DivIcon(
        icon_size=(150, 36),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12pt; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_rail),
    )
)

city_marker = folium.Marker(
    location=[city_lat, city_lon],
    icon=DivIcon(
        icon_size=(150, 36),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12pt; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_city),
    )
)

# Add markers to the map
site_map.add_child(highway_marker)
site_map.add_child(rail_marker)
site_map.add_child(city_marker)

# Draw lines between launch site and each point
coordinates_highway = [[launch_site_lat, launch_site_lon], [highway_lat, highway_lon]]
coordinates_rail = [[launch_site_lat, launch_site_lon], [rail_lat, rail_lon]]
coordinates_city = [[launch_site_lat, launch_site_lon], [city_lat, city_lon]]

line_highway = folium.PolyLine(locations=coordinates_highway, weight=1, color='blue')
line_rail = folium.PolyLine(locations=coordinates_rail, weight=1, color='green')
line_city = folium.PolyLine(locations=coordinates_city, weight=1, color='red')

# Add lines to the map
site_map.add_child(line_highway)
site_map.add_child(line_rail)
site_map.add_child(line_city)

site_map.save('map6_proximities.html')

# Display the map
site_map
```
